### Data (read from file system into dataframe and labeled)

```r
abalone_data <- read.csv("../data/abalone.data.txt", header = FALSE)
```

```r
colnames(abalone_data) <- c("Sex", "Length", "Diameter", "Height", "Whole weight", 
                            "Shucked weight", "Viscera weight", "Shell weight", "Rings")
```

### Size of the dataset

a) The dataset contains 4177 rows and 9 columns (8 features + 1 target). 

```r
dim(abalone_data)
```

b) Size of data on file system is 191873 bytes (as below)

```r
print(system("ls -la ../data | grep 'abalone.data'", intern = TRUE))
```

c) Size of data when loaded in memory

```r
object.size(abalone_data)
```

### Data Types

While the data type of attribute "Sex" is Character (which may fall into categories "F", "M", "I" (Infant)),
other attributes are of type numeric (float).

Target variable Ring has datatype integer.


```r
levels(abalone_data$Sex)

head(abalone_data)
```

### Features and Target

The features or independent variables or the predictors are **_Sex, Length, Diameter, Height, Whole Weight, Sucked Weight, Viscera Weight and Shell Weight_**.

The target or the dependent variable or the variable we want to predict is **_Rings_**
